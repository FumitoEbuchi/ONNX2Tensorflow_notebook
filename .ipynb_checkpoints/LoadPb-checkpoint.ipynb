```python
import tensorflow as tf
```

```python
def load_pb(path_to_pb):
    with tf.gfile.GFile(path_to_pb, 'rb') as f:
        graph_def = tf.GraphDef()
        graph_def.ParseFromString(f.read())
    with tf.Graph().as_default() as graph:
        tf.import_graph_def(graph_def, name='')
        return graph
```

```python
tf_graph = load_pb(f'./model.pb')
```

```python
sess = tf.Session(graph = tf_graph)
for op in tf_graph.get_operations():
    print(op.values())

output_tensor = tf_graph.get_tensor_by_name(f'test_output:0')
input_tensor = tf_graph.get_tensor_by_name(f'test_input:0')
import numpy as np
dummy_input = np.ones((1,1,128,128))
import time
start = time.time()
output = sess.run(output_tensor, feed_dict={input_tensor : dummy_input})
end = time.time()
print(output.shape)
print(f'time = {end-start}')
```
